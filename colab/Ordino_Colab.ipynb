## Streamlit’i başlat

Hücre çalışırken bekleyin. Çıktıda Colab **proxy** linki için aşağıdaki ikinci hücreyi kullanın veya Streamlit’in verdiği URL’yi deneyin.

```python
import os
import subprocess
import sys
from pathlib import Path

ROOT = Path("/content/ordino-yagci-planlama")
os.chdir(ROOT)
sys.path.insert(0, str(ROOT))

os.environ.setdefault("ORDINO_ADMIN_USER", "admin")
os.environ.setdefault("ORDINO_ADMIN_PASSWORD", "7283")

subprocess.Popen(
    [sys.executable, "-m", "streamlit", "run", "app.py", "--server.port", "8501", "--server.headless", "true"],
    cwd=str(ROOT),
)
print("Streamlit 8501’de başlatıldı (birkaç saniye bekleyin).")
```
